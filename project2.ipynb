## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

url = "https://en.wikipedia.org/wiki/BTS"
source = requests.get(url).text
soup = bs4.BeautifulSoup(source)
print(soup.get_text())
```

```python
#download an image with beautifulsoup and save it in this repository

img_tags = soup.find_all('img')
img_tags[1]['src']
```

```python
img_url = img_tags[0]['src']
img_url = 'https://wikipedia.org' + img_url
img = requests.get(img_url)
```

```python
with open('bts.jpg', 'wb') as f:
    f.write(img.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
```

```python
links = soup.find_all('a')
link_url = []
for i in links:
    try:
        link_url.append(i['href'])
    except:
        continue
print(link_url)
```

```python
link_url = [i if 'http' in i else 'https://wikipedia.org' + i for i in link_url][:10]
```

```python
link_url
```

```python
def clean(text):
    text = text.lower()
    text = ''.join([char for char in text if char.isalpha() or char == ' '])
    text = ' '.join(text.split())
    return text
```

```python
link_text = []
title = []
count = 0
for i in link_url:
    try:
        source = requests.get(i).text
        soup = bs4.BeautifulSoup(source)
        title.append(clean(soup.title.text))
        link_text.append(clean(soup.text))
        count += 1
        if count == 10:
            break
    except:
        continue
df = pd.DataFrame({'title': title, 'text': link_text})
```

```python
df
```

```python
print(df['text'].str[:100])
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
scripts
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

def get_scripts(url):
    try:
        source = requests.get(url)
        soup = bs4.BeautifulSoup(source.text)
        return clean(soup.get_text)[:1000]
    except:
        return None

scripts['script'] = scripts['link'].apply(get_scripts)
```

```python
scripts
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
scripts.to_csv('pudding_texts.csv', index=False)
```

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
url = "https://api.themoviedb.org/3/discover/movie?include_adult=false&language=en-US&api_key=23dc317cc7c25eed6f74efef918f11ce"
api_key = '23dc317cc7c25eed6f74efef918f11ce'
headers = {"accept": "application/python", "Authorization" : "Bearer" + api_key}

response = requests.get(url, headers=headers)

response.text
```

```python
dict = response.json()
movies = pd.DataFrame(dict['results'])
movies.head()
```

```python
#download the movie posters for 10 of these movies and save them to this repository
img_url = "https://image.tmdb.org/t/p/original" + movies['poster_path'][0]
img = requests.get(img_url)
```

```python
for i in range(10):
    img_url = "https://image.tmdb.org/t/p/original" + movies['poster_path'][i]
    img = requests.get(img_url)
    with open(f'movie{i}.jpg', 'wb') as f:
        f.write(img.content)
```
